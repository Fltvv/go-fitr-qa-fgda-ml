# Import libraries

```python
import os
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from mpl_toolkits import mplot3d

from scipy.signal import savgol_filter

from sklearn.preprocessing import MinMaxScaler

from sklearn.decomposition import PCA
```

```python
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', 50)
```

# Loading data and preprocessing

```python
path_to_uv_vis = './../../data/uv_vis_data/uv_vis_all.csv'
```

```python
initial_data = pd.read_csv(path_to_uv_vis, sep=';')

initial_data.head(3)
```

## Initial data

```python
# deleting row with index 0 (it contains 'Wavelength (nm)' and Abs)
data = initial_data.loc[1:, :]

# getting indexes
index = [int(col) for col in data.columns if 'Unnamed' not in col]

# getting wavelength column and absorbance columns
wavelength = pd.DataFrame(data.iloc[:, 0]).rename(columns={'1': 'wavelength'})

abosrbance_columns = [col for col in data.columns if 'Unnamed' in col]
absorbances = data[abosrbance_columns]

# renaming absorbance columns
columns = {abosrbance_columns[i]: index[i] for i in range(len(abosrbance_columns))}
absorbances = absorbances.rename(columns=columns)

# data transposition an concatenating
data = pd.concat([wavelength.T, absorbances.T], axis=0)

# converting ',' t '.' in float numbers
data = data.replace(',', '.', regex=True).astype(float)

data.head(3)
```

```python
data.to_csv('./../../data/uv_vis_data/uv_vis.csv')
```

## Data smoothing

```python
data_smoothed = data.copy()

for idx in data_smoothed.index[1:]:
    data_smoothed.loc[idx] = savgol_filter(data_smoothed.loc[idx], window_length=63, polyorder=3)

data_smoothed.head(3)
```

```python
data_smoothed.to_csv('./../../data/uv_vis_data/uv_vis_smoothed.csv')
```

## Data scaling

```python
data_scaled = data.copy()

scaler = MinMaxScaler()

data_scaled.iloc[1:] = scaler.fit_transform(data_smoothed.iloc[1:].T).T
data_scaled.head(3)
```

```python
data_scaled.to_csv('./../../data/uv_vis_data/uv_vis_scaled.csv')
```

## Calculating spectra derivatives and its visualization

```python
derivative_df_dict = {}

for i in range(0, 5):
    if i == 0:
        data_der = data_scaled.copy()
        derivative_df_dict[i] = data_der
        continue
    else:
        data_der = derivative_df_dict[i-1].copy()

    wavelength = data_der.loc['wavelength']

    for idx in data_der.index[1:]:
        data_der.loc[idx] = savgol_filter(np.gradient(data_der.loc[idx], wavelength), window_length=63, polyorder=3)

        scaler = MinMaxScaler()
        data_der.loc[idx] = scaler.fit_transform(pd.DataFrame(data_der.loc[idx])).T

    derivative_df_dict[i] = data_der
```

## Fast AdHoc

```python
temp_df = derivative_df_dict[0]
wavelength = temp_df.loc['wavelength']

x = temp_df.loc['wavelength']
rgo_spectrum = temp_df.loc[10]
go_spectrum = temp_df.loc[9]
god_spectrum = temp_df.loc[23]
```

```python
tmp = pd.concat([x, go_spectrum, god_spectrum, rgo_spectrum], axis=1)
tmp.columns = ['Длина волны, нм', 'вОГ', 'ОГ', 'ОГ_{диализ}']
# tmp.to_csv('tmp_df.csv')
```

```python
data = tmp.copy()
```

```python
data["вОГ"] = data["вОГ"].replace(0, 1e-6)
data["ОГ"] = data["ОГ"].replace(0, 1e-6)
data["ОГ_{диализ}"] = data["ОГ_{диализ}"].replace(0, 1e-6)

# Calculate log-log values
data["log(λ)"] = np.log(data["Длина волны, нм"])
data["log(вОГ)"] = np.log(data["вОГ"])
data["log(ОГ)"] = np.log(data["ОГ"])
data["log(ОГ_{диализ})"] = np.log(data["ОГ_{диализ}"])

# Plot log-log graphs for each sample
plt.figure(figsize=(10, 6))
for sample in ["вОГ", "ОГ", "ОГ_{диализ}"]:
    plt.plot(data["log(λ)"], data[f"log({sample})"], label=sample)

# Graph settings
plt.xlabel("log(λ)")
plt.ylabel("log(Intensity)")
plt.title("Log-Log Plot of Intensity vs. Wavelength")
plt.legend()
plt.grid()
plt.show()

# Calculate slopes to verify Rayleigh scattering (-4)
slopes = {}
for sample in ["вОГ", "ОГ", "ОГ_{диализ}"]:
    clean_data = data.dropna(subset=[f"log({sample})"])
    slope = np.polyfit(clean_data["log(λ)"], clean_data[f"log({sample})"], 1)[0]
    slopes[sample] = slope

slopes
```

```python
filtered_data = data[(data["log(λ)"] >= 5.6) & (data["log(λ)"] <= 6.4)]

# Calculate slopes (tangents of the angle) for the specified range
tangents = {}
for sample in ["вОГ", "ОГ", "ОГ_{диализ}"]:
    # Ensure there are no NaN values
    clean_data = filtered_data.dropna(subset=[f"log({sample})"])
    if len(clean_data) > 1:  # Ensure there are enough points for a fit
        tangent = np.polyfit(clean_data["log(λ)"], clean_data[f"log({sample})"], 1)[0]
        tangents[sample] = tangent

tangents
```

```python
fontsize = 14

for i in range(0, 5):
    temp_df = derivative_df_dict[i]
    wavelength = temp_df.loc['wavelength']

    x = temp_df.loc['wavelength']
    rgo_spectrum = temp_df.loc[10]
    go_spectrum = temp_df.loc[9]
    god_spectrum = temp_df.loc[23]

    sns.set_theme()

    fig = plt.figure(figsize=(16, 4))
    ax = fig.gca()

    ax.plot(x, rgo_spectrum, color='orange', label='$вОГ$ $(id = 10)$')
    ax.plot(x, go_spectrum, color='dodgerblue', label='$ОГ$ $(id = 9)$')
    ax.plot(x, god_spectrum, color='green', label='$ОГ_{диализ}$ $(id = 23)$')

    ax.set_xlabel('Длина волны, нм', fontstyle='normal', fontsize=fontsize, labelpad=5)
    ax.set_ylabel('$A_{n}$', fontstyle='italic', fontsize=fontsize, labelpad=15, rotation=0)

    ax.legend(loc='best')

    plt.tight_layout()

    plt.savefig(path_to_save + f'/3_curves_derivative_{i}.png', dpi=600)
```

## Dimensionality Reduction *(PCA)*

```python
fontsize = 14
fontstyle='normal'
markersize = 256

random_state = 42
n_components = 2

for i in range(0, 5):
    temp_df = derivative_df_dict[i]
    spectral_data = temp_df.iloc[1:, :]

    pca = PCA(n_components=n_components, random_state=random_state)
    spectral_data_pca = pd.DataFrame(pca.fit_transform(spectral_data), columns=[f'PC_{i}' for i in range(1, n_components+1)])
    
    explained_variance = [round(k, 4) for k in pca.explained_variance_ratio_]

    spectral_data_pca = pd.concat([spectral_data_pca, samples_description['Type']], axis=1)

    sns.set_theme()

    fig = plt.figure(figsize=(8, 3.5))
    ax = fig.gca()

    sns.scatterplot(data=spectral_data_pca,
                    x='PC_1',
                    y='PC_2',
                    hue='Type',
                    palette='Dark2',
                    s=markersize)

    ax.set_xlabel(f'Principal component 1 \n (PoEV = {explained_variance[0]})', fontstyle=fontstyle, fontsize=fontsize, labelpad=5)
    ax.set_ylabel(f'Principal component 2 \n (POEV = {explained_variance[1]})', fontstyle=fontstyle, fontsize=fontsize, labelpad=5)

    ax.legend(loc='best')

    plt.tight_layout()

    plt.savefig(path_to_save + f'/pca_derivative_{i}.png', dpi=600)

    plt.show()
```
