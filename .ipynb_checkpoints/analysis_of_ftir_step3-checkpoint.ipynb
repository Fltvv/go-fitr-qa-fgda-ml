```python
path_ir = 'data/ftir'

path_to_curves = os.listdir(path_ir)
path_to_curves = sorted(path_to_curves, key=lambda x: int(x.split('.')[0]))
path_to_curves = [path_ir + '/' + filename for filename in path_to_curves]
path_to_curves
```

```python
ir_id = {}

for i, p in enumerate(path_to_curves):
    name = p.split('/')[-1].split('.')[0]
    ir_id[i] = name
ir_id
```

```python
def extract_ir_data(path):
    with open(path, 'r') as f:
        data = f.readlines()
        x_lst, y_lst = [], []
        for line in data:
            line = line.rstrip()
            x, y = map(float, line.split(','))
            x_lst.append(x)
            y_lst.append(y)
    return x_lst, y_lst
```

```python
full_X = extract_ir_data(path_to_curves[0])[0]
X = list(filter(lambda x: x >= 800, full_X))
```

```python
Y = [extract_ir_data(p)[1][:len(X)] for p in path_to_curves]  # initial
```

```python
Y_smooth = [savgol_filter(y, window_length=53, polyorder=3) for y in Y]  # smoothed
```

```python
scaler = MinMaxScaler()
Y_smooth = scaler.fit_transform(np.array(Y_smooth).T).T  # smoothed and normalized
```

```python
Y_torch = torch.Tensor([Y_smooth])
Y_torch.shape
```

```python
Y_torch = Y_torch.permute(1, 0, 2)
Y_torch.shape
```

```python
def encoder_block(in_channels, out_channels, kernel_size, padding):
    block = nn.Sequential(
        nn.Conv1d(in_channels, out_channels, kernel_size=kernel_size, padding=padding),
        nn.Sigmoid(),
        nn.MaxPool1d(kernel_size=2)
    )
    return block

def decoder_block(in_channels, out_channels, kernel_size, padding):
    block = nn.Sequential(
        nn.Conv1d(in_channels, out_channels, kernel_size=kernel_size, padding=padding),
        nn.Sigmoid(),
        nn.Upsample(scale_factor=2, mode='linear')
    )
    return block

class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()
        
        '''142'''
        self.encoder = nn.Sequential(
#             encoder_block(in_channels=1, out_channels=2, kernel_size=3, padding=1), # 1142 --> 1142 --> 571
#             encoder_block(in_channels=2, out_channels=4, kernel_size=2, padding=0), # 571 --> 570 --> 285
#             encoder_block(in_channels=4, out_channels=8, kernel_size=2, padding=0) # 285 --> 284 --> 142
#         )
        
#         self.decoder = nn.Sequential(
#             decoder_block(in_channels=8, out_channels=4, kernel_size=2, padding=1), # 142 --> 143 --> 286
#             decoder_block(in_channels=4, out_channels=2, kernel_size=2, padding=0), # 286 --> 285 --> 570
#             decoder_block(in_channels=2, out_channels=1, kernel_size=2, padding=1) # 570 --> 571 --> 1142
#         )
        
        '''71'''
        self.encoder = nn.Sequential(
            encoder_block(in_channels=1, out_channels=2, kernel_size=3, padding=1), # 1142 --> 1142 --> 571
            encoder_block(in_channels=2, out_channels=4, kernel_size=2, padding=0), # 571 --> 570 --> 285
            encoder_block(in_channels=4, out_channels=8, kernel_size=2, padding=0), # 285 --> 284 --> 142
            encoder_block(in_channels=8, out_channels=16, kernel_size=3, padding=1) # 142 --> 142 --> 71
        )

        self.decoder = nn.Sequential(
            decoder_block(in_channels=16, out_channels=8, kernel_size=3, padding=1), # 71 --> 71 --> 142
            decoder_block(in_channels=8, out_channels=4, kernel_size=2, padding=1), # 142 --> 143 --> 286
            decoder_block(in_channels=4, out_channels=2, kernel_size=2, padding=0), # 286 --> 285 --> 570
            decoder_block(in_channels=2, out_channels=1, kernel_size=2, padding=1) # 570 --> 571 --> 1142
        )

#         '''35'''
#         self.encoder = nn.Sequential(
#             encoder_block(in_channels=1, out_channels=2, kernel_size=3, padding=1), # 1142 --> 1142 --> 571
#             encoder_block(in_channels=2, out_channels=4, kernel_size=2, padding=0), # 571 --> 570 --> 285
#             encoder_block(in_channels=4, out_channels=8, kernel_size=2, padding=0), # 285 --> 284 --> 142
#             encoder_block(in_channels=8, out_channels=16, kernel_size=3, padding=1), # 142 --> 142 --> 71
#             encoder_block(in_channels=16, out_channels=32, kernel_size=2, padding=0) # 71 --> 70 --> 35
#         )
        
#         self.decoder = nn.Sequential(
#             decoder_block(in_channels=32, out_channels=16, kernel_size=3, padding=1), # 35 --> 35 --> 70
#             decoder_block(in_channels=16, out_channels=8, kernel_size=2, padding=1), # 70 --> 71 --> 142
#             decoder_block(in_channels=8, out_channels=4, kernel_size=2, padding=1), # 142 --> 143 --> 286
#             decoder_block(in_channels=4, out_channels=2, kernel_size=2, padding=0), # 286 --> 285 --> 570
#             decoder_block(in_channels=2, out_channels=1, kernel_size=2, padding=1) # 570 --> 571 --> 1142
#         )

    def forward(self, x):
        latent = self.encoder(x)
        reconstruction = self.decoder(latent)
        return reconstruction, latent
```

```python
outputs = []

def train(model, loss_function, optimizer, n_epochs=1000):
    epochs = []
    losses = []
    for epoch in range(1, n_epochs+1):
        reconstructed, _ = model(Y_torch)
        loss = loss_function(reconstructed, Y_torch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        losses.append(loss.item())
        epochs.append(epoch)

        if epoch % 50 == 0:
            clear_output(True)

            fig = plt.figure(figsize=(13, 5), dpi=150, linewidth=1.0)
            plt.grid(True)
            plt.tick_params(direction='in')
            ax = fig.gca()
            ax.scatter(epochs,
                       losses,
                       marker='s',
                       color='blue',
                       edgecolor='black',
                       alpha=0.6,
                       label='points of local maximum')
            ax.set_xlabel('number of epoch', fontname='DejaVu Sans', fontstyle='italic', fontsize=14)
            ax.set_ylabel('$MSE$', fontname='DejaVu Sans', fontsize=14, rotation=90, labelpad=15)
            plt.show()
```

```python
autoencoder = Autoencoder()
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(autoencoder.parameters(), lr=5e-3)
```

```python
train(model=autoencoder, loss_function=loss_function, optimizer=optimizer, n_epochs=1000)
```

# 4. Кластеризация на эмбеддингах

```python
def plot_metrics(data, max_clusters=10, metric='euclidean', save='n', name_of_file='f', style='default', dpi=300):
    inertia = []
    silhouette = []
    clusters_range = range(1, max_clusters)

    for k in tqdm(clusters_range):
        kmeans = KMeans(n_clusters=k,
                        max_iter=100_000,
                        random_state=42)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
        if k > 1:
            silhouette.append(silhouette_score(data, kmeans.labels_))

    fig, ax = plt.subplot_mosaic([['left', 'right']],
                                 gridspec_kw={'width_ratios': [1, 1]},
                                 figsize=(16, 5),
                                 dpi=dpi,
                                 layout='constrained')

    ax['left'].grid(True)
    ax['left'].tick_params(direction='in')
    ax['left'].set_title(f'Elbow method',
                         fontname='DejaVu Sans',
                         fontsize=20,
                         fontstyle='italic',
                         fontweight='heavy')
    ax['left'].plot(clusters_range,
                    inertia,
                    '-o',
                    linewidth=2.0,
                    color='orange')
    ax['left'].set_xlabel('$number$ $of$ $clusters$',
                          fontname='DejaVu Sans',
                          fontsize=16)
    ax['left'].set_ylabel('$WCSS$',
                          fontname='DejaVu Sans',
                          fontsize=16)

    ax['right'].grid(True)
    ax['right'].tick_params(direction='in')
    ax['right'].set_title(f'Silhouette score',
                          fontname='DejaVu Sans',
                          fontsize=20,
                          fontstyle='italic',
                          fontweight='heavy')
    ax['right'].plot(clusters_range[1:],
                     silhouette,
                     '-o',
                     linewidth=2.0,
                     color='orange')
    ax['right'].set_xlabel('$number$ $of$ $clusters$',
                           fontname='DejaVu Sans',
                           fontsize=16)
    ax['right'].set_ylabel('$SS$',
                           fontname='DejaVu Sans',
                           fontsize=16)

    fig.tight_layout()
    if save in ['y', 'Y', 'yes']:
        plt.savefig(name_of_file, dpi=dpi)
    plt.show()
```

```python
def cluster_analysis(data, n_clusters):
    info = pd.DataFrame(data.iloc[:, 0])
    clust_data = data.iloc[:, 1:]
    kmeans = KMeans(n_clusters=n_clusters,
                    init='k-means++',
                    random_state=42)
    groups = kmeans.fit_predict(clust_data)
    info.insert(loc=0, column='cluster_label', value=groups)
    info['cluster_label'] += 1
    info = info.sort_values(by='cluster_label')
    return info
```
