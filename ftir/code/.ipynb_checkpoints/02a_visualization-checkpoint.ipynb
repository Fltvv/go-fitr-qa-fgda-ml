# Import libraries

```python
import os
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from mpl_toolkits import mplot3d
```

```python
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', 50)
```

```python
path_to_save = './../results/visualization'
```

# Loading data

## Train data (*smoothed & scaled*)

```python
path_to_ftir_data = './../data/preprocessed_data/ftir_scaled_train.csv'
```

```python
data = pd.read_csv(path_to_ftir_data, index_col=0)

print(data.shape)
data.head(3)
```

## Samples description

```python
path_to_samples_description = './../data/preprocessed_data/ftir_description.csv'
```

```python
samples_description = pd.read_csv(path_to_samples_description, index_col=0)

samples_description.index = samples_description.index.astype(str)

print(samples_description.shape)
samples_description.head(3)
```

# Visualization

## Selecting *train* part of sample `samples_description` 

```python
samples_description = samples_description.loc[samples_description['Label'] == 'Train']

print(samples_description.shape)
samples_description.head(3)
```

## Smoothed and scaled spectra for each class (3D)

```python
type_color_dict = {
    '$GO$': 'dodgerblue',
    '$rGO$': 'orange',
    '$GO_{d}$': 'green'
}

wavenumber = data.loc['wavenumber']
```

```python
fontsize = 14

sns.set_theme()

fig = plt.figure(figsize=(12, 12), dpi=300)
ax = plt.axes(projection='3d')

for index in data.index[1:]:
    type = samples_description.loc[index]['Type']
    color = type_color_dict[type]

    x = wavenumber
    y = [int(index)] * data.shape[1]
    z = data.loc[index]
    
    ax.plot(x, y, z, color=color, linewidth=1.5, label=type)

ax.set_xlabel('Wavenumber, cm$^{-1}$', fontstyle='normal', fontsize=fontsize, labelpad=15)
ax.set_ylabel('Sample ID', fontstyle='normal', fontsize=fontsize, labelpad=15)
ax.set_zlabel('A', fontstyle='italic', fontsize=fontsize, labelpad=5, rotation=0)

# only unique labels
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc='center right')

plt.tight_layout()

ax.view_init(elev=15, azim=30, roll=0)

# plt.savefig(f'{path_to_save}/3d_spectra_plot.png', dpi=600)

plt.show()
```

## Principal 3 spectra (2D)

```python
x = data.loc['wavenumber']

rgo_spectrum = data.loc['10']
go_spectrum = data.loc['8']
god_spectrum = data.loc['22']
```

```python
fontsize = 16

sns.set_theme()

fig = plt.figure(figsize=(10, 7))
ax = fig.gca()

ax.plot(x, rgo_spectrum, color='orange', label='$вОГ$ $(id = 10)$')
ax.plot(x, go_spectrum, color='dodgerblue', label='$ОГ$ $(id = 8)$')
ax.plot(x, god_spectrum, color='green', label='$ОГ_{диализ}$ $(id = 20)$')

ax.set_xlabel('Волновое число, см$^{-1}$', fontstyle='normal', fontsize=fontsize, labelpad=5)
ax.set_ylabel('$A_{n}$', fontstyle='italic', fontsize=fontsize, labelpad=15, rotation=0)

ax.invert_xaxis()

ax.legend(loc='best')

plt.tight_layout()

plt.savefig(f'{path_to_save}/2d_princripal_spectra.png', dpi=600)

plt.show()
```

## Principal spectra of each class — 3 subplots

```python
fontsize = 16
fontstyle='italic'

wavenumber = data.loc['wavenumber']
unique_types = samples_description['Type'].unique()
```

```python
for type in unique_types:
    type_indexes = samples_description.loc[samples_description['Type'] == type].index
    hue = samples_description.loc[type_indexes]
    data_type = data.loc[type_indexes]

    color = type_color_dict[type]

    sns.set_theme()
    
    fig = plt.figure(figsize=(20, 4))
    ax = fig.gca()

    for i in type_indexes:
        spectrum = data_type.loc[i]
        ax.plot(
            wavenumber,
            spectrum,
            color=color,
            linewidth=2.0,
            label=type
        )
    
    ax.set_xlabel(
        'Волновое число, см$^{-1}$',
        fontstyle='normal',
        fontsize=fontsize,
        labelpad=5
    )
    ax.set_ylabel(
        '$A_{n}$',
        fontstyle='italic',
        fontsize=fontsize,
        labelpad=15,
        rotation=0
    )
    
    ax.invert_xaxis()
    
    # only unique labels
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    # ax.legend(by_label.values(), by_label.keys(), loc='best')
    
    plt.tight_layout()
    
    plt.savefig(f'{path_to_save}/princripal_spectra_{type}.png', dpi=600)
    
    plt.show()
```

## Heatmap visualization

```python
# heatmap data processing
heatmap_data = data.copy().loc['1':]
heatmap_data.columns = [round(wave) for wave in data.loc['wavenumber'].tolist()]
heatmap_data = heatmap_data.loc[:, 1900:]

# sorting indexed (GO, rGO, GOd)
indexes = []
unique_types = sorted(samples_description['Type'].unique())
for type in unique_types:
    type_indexes = samples_description.loc[samples_description['Type'] == type].index
    indexes.extend(type_indexes)

# reindexing heatmap dataframe
heatmap_data = heatmap_data.reindex(indexes)
heatmap_data.head(3)
```

### All samples

```python
fontsize = 12
fontstyle='normal'

sns.set_theme()

fig = plt.figure(figsize=(16, 5), dpi=150, linewidth=1.0)
ax = fig.gca()

# cmap = ['YlGnBu', 'PiYG', 'Reds', 'BuPu', 'rocket', 'flare', 'cubehelix', 'vlag', 'Spectral', 'icefire', 'Blues']
heatmap = sns.heatmap(
    data=heatmap_data,
    cmap='Reds',
    ax=ax)

ax.set_ylabel(
    'Sample ID',
    fontname='DejaVu Sans',
    fontsize=fontsize,
    fontstyle=fontstyle,
    labelpad=15
)
ax.set_xlabel(
    'Wavenumber, cm$^{-1}$',
    fontstyle=fontstyle,
    fontsize=fontsize,
    labelpad=5
)

ax.set_yticklabels(heatmap.get_yticklabels(), rotation=0, fontsize=12)

plt.tight_layout()
  
# plt.savefig(f'{path_to_save}/heatmap.png', dpi=600)

plt.show()
```

### Grouped by sample type

```python
fontsize = 12
fontstyle='normal'

unique_types = samples_description['Type'].unique()

type_cmap = {
    '$GO$': 'Blues',
    '$GO_{d}$': 'Greens',
    '$rGO$': 'Oranges'
}

for type in unique_types:
    type_indexes = samples_description.loc[samples_description['Type'] == type].index
    temp_data = heatmap_data.loc[type_indexes]

    sns.set_theme()
    
    fig = plt.figure(figsize=(20, 4.5))
    ax = fig.gca()

    # cmap = ['YlGnBu', 'PiYG', 'Reds', 'BuPu', 'rocket', 'flare', 'cubehelix', 'vlag', 'Spectral', 'icefire', 'Blues']
    cmap = type_cmap[type]
    
    heatmap = sns.heatmap(
        data=temp_data,
        cmap=cmap,
        ax=ax
    )
    
    ax.set_xlabel(
        'Волновое число, см$^{-1}$',
        fontstyle=fontstyle,
        fontsize=fontsize,
        labelpad=5
    )
    ax.set_ylabel(
        'ID образца',
        fontname='DejaVu Sans',
        fontsize=fontsize,
        fontstyle=fontstyle,
        labelpad=5
    )

    ax.set_yticklabels(heatmap.get_yticklabels(), rotation=0, fontsize=12)

    plt.tight_layout()
    
    plt.savefig(f'{path_to_save}/heatmap_{type}.png', dpi=600)        
              
    plt.show()
```

## Analysis of spectra reproducibility

```python
reproducibility_data = heatmap_data.copy()

print(reproducibility_data.shape)
reproducibility_data.head(3)
```

### *1 plot at 1 figure*

```python
fontsize = 16
fontstyle='normal'

unique_types = samples_description['Type'].unique()

type_color = {
    '$GO$': 'blue',
    '$GO_{d}$': 'green',
    '$rGO$': 'orange'
}

for type in unique_types:
    type_indexes = samples_description.loc[samples_description['Type'] == type].index
    temp_data = reproducibility_data.loc[type_indexes]

    repr_arr = round(temp_data.std() / temp_data.mean() * 100, 2)
    repr_df = pd.DataFrame(repr_arr, columns=['rsd'])

    sns.set_theme()

    fig = plt.figure(figsize=(20, 4.5))
    ax = fig.gca()

    color = type_color[type]
    
    sns.lineplot(
        data=repr_df,
        x=repr_df.index,
        y='rsd',
        color=color,
        linewidth=2.0,
        # label=type
        ax=ax
    )
    
    ax.set_xlabel(
        'Волновое число, см$^{-1}$',
        fontstyle='normal',
        fontsize=fontsize,
        labelpad=5,
    )
    
    ax.set_ylabel(
        '$s_{r}$, %',
        fontstyle='italic',
        fontsize=fontsize,
        labelpad=25,
        rotation=0,
    )

    ax.invert_xaxis()

    plt.tight_layout()
    
    plt.savefig(f'{path_to_save}/rsd_visualization_{type}.png', dpi=600)        
              
    plt.show()
```

### *3 plots at 1 figure*

```python
fontsize = 22
fontstyle='normal'

unique_types = samples_description['Type'].unique()

type_color = {
    '$GO$': 'blue',
    '$GO_{d}$': 'green',
    '$rGO$': 'orange'
}

label_mapper_dict = {
    '$GO$': '$ОГ$',
    '$GO_{d}$': '$ОГ_{диализ}$',
    '$rGO$': '$вОГ$'
}

fig = plt.figure(figsize=(16, 10))
ax = fig.gca()

for type in unique_types:
    type_indexes = samples_description.loc[samples_description['Type'] == type].index
    temp_data = reproducibility_data.loc[type_indexes]

    repr_arr = round(temp_data.std() / temp_data.mean() * 100, 2)
    repr_df = pd.DataFrame(repr_arr, columns=['rsd'])

    sns.set_theme()

    color = type_color[type]
    label = label_mapper_dict[type]
    
    sns.lineplot(
        data=repr_df,
        x=repr_df.index,
        y='rsd',
        color=color,
        linewidth=2.0,
        label=label,
        ax=ax
    )
    
    ax.set_xlabel(
        'Волновое число, см$^{-1}$',
        fontstyle='normal',
        fontsize=fontsize,
        labelpad=5,
    )
    
    ax.set_ylabel(
        '$s_{r}$, %',
        fontstyle='italic',
        fontsize=fontsize,
        labelpad=5,
        rotation=90,
    )
    ax.tick_params(axis='both', which='major', labelsize=fontsize-2)

    ax.set_yscale('log')
    
    ax.invert_xaxis()

    plt.tight_layout()

ax.legend(fontsize=18)

plt.savefig(f'{path_to_save}/rsd_visualization_log_scale.png', dpi=600)        
          
plt.show()
```

*Same plot but in `ln` scale*

```python
fontsize = 22
fontstyle='normal'

unique_types = samples_description['Type'].unique()

type_color = {
    '$GO$': 'blue',
    '$GO_{d}$': 'green',
    '$rGO$': 'orange'
}

label_mapper_dict = {
    '$GO$': '$ОГ$',
    '$GO_{d}$': '$ОГ_{диализ}$',
    '$rGO$': '$вОГ$'
}

fig = plt.figure(figsize=(16, 10))
ax = fig.gca()

for type in unique_types:
    type_indexes = samples_description.loc[samples_description['Type'] == type].index
    temp_data = reproducibility_data.loc[type_indexes]

    repr_arr = round(temp_data.std() / temp_data.mean(), 4)
    repr_df = pd.DataFrame(repr_arr, columns=['rsd'])

    repr_df['rsd'] = repr_df['rsd'].apply(lambda x: np.log(x))

    sns.set_theme()

    color = type_color[type]
    label = label_mapper_dict[type]
    
    sns.lineplot(
        data=repr_df,
        x=repr_df.index,
        y='rsd',
        color=color,
        linewidth=2.0,
        label=label,
        ax=ax
    )
    
    ax.set_xlabel(
        'Волновое число, см$^{-1}$',
        fontstyle='normal',
        fontsize=fontsize,
        labelpad=5,
    )
    
    ax.set_ylabel(
        '$ln(s_{r})$',
        fontstyle='italic',
        fontsize=fontsize,
        labelpad=5,
        rotation=90,
    )
    ax.tick_params(axis='both', which='major', labelsize=fontsize-2)
    
    ax.invert_xaxis()

    plt.tight_layout()

ax.legend(fontsize=18)

# plt.savefig(f'{path_to_save}/rsd_visualization_log_scale.png', dpi=600)        
          
plt.show()
```
